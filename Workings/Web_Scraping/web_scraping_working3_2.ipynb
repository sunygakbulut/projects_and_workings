```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

## 1. Go to your favorite actress'/actor's imdb page

```python
#Tom Cruise
base_site = "https://www.imdb.com/filmosearch/?sort=num_votes&explore=title_type&role=nm0000129&ref_=nm_flmg_shw_4"

response = requests.get(base_site)
response.status_code
```

```python
html = response.content
soup = BeautifulSoup(html, "lxml")
```

## 2. Get all the links to his person's movies

```python
raw = soup.find_all("h3",{"class":"lister-item-header"})
```

```python
len(raw)
```

```python
raw[0]
```

```python
raw[0].find("a").text
```

```python
links = ["https://www.imdb.com/"+i.find("a").get("href")+"?ref_=filmo_li_tt" for i in raw]
```

## 3. Get each movie's name, director and imdb score

### Names

```python
names = [i.find("a").text for i in raw]
```

```python
names_df = pd.DataFrame(names, columns = ["Names"])
```

```python
names_df.head()
```

### Directors

```python
raw1 = soup.find_all("div",{"class":"lister-item-content"})
```

```python
raw1[0].find_all("p")[2].find_all("a")[0].text
```

```python
director = [i.find_all("p")[2].find_all("a")[0].text for i in raw1]
```

```python
director_df = pd.DataFrame(director, columns = ["Directors"])
```

```python
director_df.head()
```

### Metascores

```python
raw2 = soup.find_all("div",{"class":"inline-block ratings-metascore"})
```

```python
raw2[0]
```

```python
raw2[0].find("span").text.strip()
```

```python
score = [i.find("span").text.strip() for i in raw2]
```

```python
score_df = pd.DataFrame(score, columns = ["Scores"])
```

### Combine

```python
combine = pd.concat([names_df, director_df, score_df], axis = 1)
combine
```

```python
combine = combine[["Scores", "Names", "Directors"]]
```

```python
combine = combine.sort_values(by = ("Names"), ascending = False).reset_index(drop=True)
```
