```python
soup = BeautifulSoup(html, 'lxml')
soup.find_all("table")
```

```python
sonuc = pd.DataFrame()
type(sonuc)
```

```python
sonuc = pd.DataFrame()
for k in range(len(country_text)):
    table = soup.find_all("table")[k+1]
    tables = table.find_all('tr')
    name_text = []
    population_text = []
    for i in tables:
        name_text.append(i.contents[1].text.strip())
        population_text.append(i.contents[5].text.strip())

    popu = pd.DataFrame(population_text[1:], columns = ["Population"])
    nm = pd.DataFrame(name_text[1:], columns = ["Names"])
    xx = pd.concat([nm,popu], axis = 1)
    xx["Country_Name"] = country_text[k]
    sonuc = pd.concat([sonuc,xx], axis = 0)

```
