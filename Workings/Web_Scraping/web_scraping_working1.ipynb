## Import Library

```python
import requests
from bs4 import BeautifulSoup
```

## Making a get request

```python
base_site = 'https://en.wikipedia.org/wiki/Big_data'
response = requests.get(base_site)
response
```

```python
html = response.content
html;
```

### Making the soup

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
type(soup)
```

```python
with open('Bigdata_response.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
soup;
```

## Exercise:1

```python
a = soup.find_all("a")
```

```python
a
```

```python
clean_links = [l for l in a if l.get('href') != None]
clean_links
```

## Exercise:2

```python
h2 = soup.find_all("h2")
```

```python
extracted_h2 = [l.text for l in h2]
extracted_h2
```

## Exercise:3

```python
base_site = 'https://en.wikipedia.org/wiki/Artificial_intelligence'
response = requests.get(base_site)
response
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
images = soup.find_all("img")
```

```python
images
```

```python
len(images)
```

```python
images[0].parent
```

```python
images[0].parent.get("href")
```

```python
images[0].parent.get("src")
```

```python
extracted_images = [k.parent.get("href") for k in images]
extracted_images
```

```python
len(extracted_images)
```

```python
clean_images = [z for z in extracted_images if z != None]
clean_images
```

```python
from urllib.parse import urljoin
```

```python
base_site
```

```python
clean_images[1]
```

```python
full_url = [urljoin(base_site, z) for z in clean_images]
full_url
```
